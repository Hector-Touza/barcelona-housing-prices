```python
# imports
import pandas as pd
from pathlib import Path
```

```python
data_raw = pd.read_json("../data_raw/data_barcelona_gracia_15_05_20.json")
```

## surface ##

```python
surface = data_raw["features"]

for house in surface:
    # surface info is almost always the first entry
    if house[0].find("útiles") is not -1:
        print(house[0].split(",")[1])
    elif house[0].find("construidos") is not -1:
        print(house[0].split(",")[0])
    
    # if it's not the first entry, search location and return value
    else:
        for feature in house[1:]:
            if feature.find("útiles") is not -1:
                print(feature.split(",")[1])
                break
            elif feature.find("construidos") is not -1:
                print(feature.split(",")[0])
                break
```
